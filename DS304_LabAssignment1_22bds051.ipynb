**IRIS Dataset**

```python
#Necessary Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
##Downloading the dataset
from sklearn.datasets import load_iris
iris = load_iris()
data = pd.DataFrame(data=iris.data, columns=iris.feature_names)
data['Species'] = iris.target
```

```python
# Displaying the first 5 rows
data.head(5)
```

```python
# Displaying the last 5 rows
data.tail(5)
```

```python
# Displaying the shape of the dataset
dataset_shape = data.shape
print("\nShape of the dataset (rows, columns):", dataset_shape)
```

```python
# Printing the column names
column_names = data.columns
print("\nColumn names:")
print(column_names)
```

```python
# Accessing the individual element in the first row and first column
element = data.iloc[0, 0]
print("\nIndividual element (1st row, 1st column):", element)
```

```python
# Getting information about the data types of each column
data_types = data.dtypes
print("\nData types of each column:")
print(data_types)
```

```python
# Adding a new column for Petal Area (product of Petal Length and Petal Width)
data['Petal Area (cm^2)'] = data['petal length (cm)'] * data['petal width (cm)']

# Displaying the dataset after adding the new column 'Petal Area'
print("\nDataset after adding a new column 'Petal Area (cm^2)':")
print(data.head())
```

```python
# Removing the 'Petal Area (cm^2)' column
data = data.drop(columns=['Petal Area (cm^2)'])

# Displaying the dataset after removing the 'Petal Area (cm^2)' column
print("\nDataset after removing the 'Petal Area (cm^2)' column:")
print(data.head())
```

```python
# Removing the first row
data = data.drop(index=0)
print("\nDataset after removing the first row:")
print(data.head())
```

```python
# Filtering the dataset where Species is "setosa"
setosa_df = data[data['Species'] == 'setosa']
print("\nFiltered dataset (only 'setosa' species):")
print(setosa_df.head())
```

```python
# Calculating summary statistics for numerical columns
summary_statistics = data.describe().T

# Mean
mean_values = data.mean()

# Median
median_values = data.median()

# Standard Deviation
std_dev_values = data.std()

# Variance
variance_values = data.var()

# Min and Max
min_values = data.min()
max_values = data.max()

# Quartiles (25%, 50%, 75%)
quartiles = data.quantile([0.25, 0.5, 0.75])

# Range (Max - Min)
range_values = max_values - min_values

# Displaying all statistics
print("Summary Statistics:\n", summary_statistics)
print("\nMean Values:\n", mean_values)
print("\nMedian Values:\n", median_values)
print("\nStandard Deviation:\n", std_dev_values)
print("\nVariance:\n", variance_values)
print("\nMin Values:\n", min_values)
print("\nMax Values:\n", max_values)
print("\nQuartiles:\n", quartiles)
print("\nRange Values:\n", range_values)
```

```python
# Counting the number of instances for each species
species_counts = data['Species'].value_counts()
print("\nNumber of instances for each species:")
print(species_counts)
```

```python
# Plot the basic bar plot for species counts
plt.figure(figsize=(8, 6))
species_counts.plot(kind='bar', color='skyblue')
plt.title('Count of Each Species in the Iris Dataset')
plt.xlabel('Species')
plt.ylabel('Count')
plt.xticks(rotation=0)
plt.show()
```
